```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('students_data.csv')
```

```python
df.head()
```

```python
expression_counts = df['Expression'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(expression_counts, labels=expression_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Expressions')
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count the number of unique persons
total_persons = df['Name'].nunique()

# Group by Expression and count the unique persons for each expression
expression_counts = df.groupby('Expression')['Name'].nunique()

# Plot the bar chart
plt.bar(expression_counts.index, expression_counts.values)

# Label axes and add title
plt.xlabel('Expression')
plt.ylabel('Number of Persons')
plt.title('Number of Persons by Expression')

# Show plot
plt.show()
```

```python
# Count the number of unique persons
total_persons = df['Name'].nunique()

# Group by Expression and count the unique persons for each expression
expression_counts = df.groupby('Expression')['Name'].nunique()

# Plot the bar chart
expression_counts.plot(kind='bar')

# Label axes and add title
plt.xlabel('Expression')
plt.ylabel('Number of Persons')
plt.title('Number of Persons by Expression')

# Show plot
plt.show()
```

```python


# Define the order of expressions
expression_order = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']

# Count the number of unique persons
total_persons = df['Name'].nunique()

# Group by Expression and count the unique persons for each expression
expression_counts = df.groupby('Expression')['Name'].nunique()

# Plot the bar chart
expression_counts.reindex(expression_order).plot(kind='bar', color='skyblue')

# Label axes and add title
plt.xlabel('Expression')
plt.ylabel('Number of Persons')
plt.title('Number of Persons by Expression')

# Set x-axis ticks to expression labels
plt.xticks(range(len(expression_order)), expression_order)

# Show plot
plt.show()
```
